```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
%matplotlib inline
```

```python
#Load the dataset and view sample data
iris = datasets.load_iris()
x = iris.data
y = iris.target
x[:5],y[:5]
```

```python
#perform 3D transform for data using PCA
pca = decomposition.PCA(n_components=3)
pca.fit(x)
x_new = pca.transform(x)
x_new[:5]
```

```python
#Prepare & Plot data in 3Dplot
fig = plt.figure(1, figsize=(5, 5))
ax = Axes3D(fig, rect=[0, 0, 1, 1],elev=35, azim=125)
for name, label in [('Sentosa', 0), ('Versicolor', 1), ('Virginica', 2)]:
    ax.text3D(x_new[y == label, 0].mean(),
              x_new[y == label, 1].mean() + 1,
              x_new[y == label, 2].mean(), name,
              horizontalalignment='center',
              bbox=dict(alpha=.3, edgecolor='w', facecolor='w'))
ax.scatter(x_new[:, 0], x_new[:, 1], x_new[:, 2], c=y, cmap=plt.cm.rainbow,
           edgecolor='g')
plt.show(ax)
```
